## Operação de adição e subtração
A operação de adição consiste em somar valores dos pixels de uma imagem com outra. O resultado é uma nova imagem.

Imports

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

Lendo as imagens

```python
path1 = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\001-pre-processing\images\Superman.png'
img1 = cv.imread(path1, 0)
```

```python
path2 = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\001-pre-processing\images\batman.png'
img2 = cv.imread(path2, 0) # O número "0" é para ler em tons de cinza
```

### Adição

Função add(img1, img2)

Entrada:
            
            1. Imagem 1 e 2
         
Saída:

            1. Matriz com valores dos pixels somados

```python
imgSomada = cv.add(img1,img2)
```

```python
fig = plt.figure(figsize=(20,5))
plt.imshow(imgSomada, cmap=plt.cm.gray) # cmap para ver em tons de cinza
```

### Subtração

```python
fig = plt.figure(figsize=(20,5))

ax1=fig.add_subplot(121)
plt.imshow(img1, cmap=plt.cm.gray)

ax2=fig.add_subplot(122)
plt.imshow(img2, cmap=plt.cm.gray)
plt.show()
```

Função subtract(imag1, imag2)

Entrada:
            
        1. Imagens 1 e 2
Saída:

        1. Matriz com os valores dos pixels somados

```python
imgSub = cv.subtract(img1, img2)
```

```python
fig = plt.figure(figsize=(20,5))
plt.imshow(imgSub, cmap=plt.cm.gray)
```
