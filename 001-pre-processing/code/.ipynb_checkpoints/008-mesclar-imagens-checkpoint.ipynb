## Operação de mistura
A operação de mistura consiste em mesclar, com perda de dados, as informações de duas imagens em uma.

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

Lendo as imagens

```python
path1 = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\001-pre-processing\images\Superman.png'
img1 = cv.imread(path1, 0)

path2 = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\001-pre-processing\images\batman.png'
img2 = cv.imread(path2, 0)

fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(121)
plt.imshow(img1, cmap=plt.cm.gray)

ax2 = fig.add_subplot(122)
plt.imshow(img2, cmap=plt.cm.gray)

plt.show()
```

Função addWeighted(src1, alpha, src2, beta, gamma)

Entrada:

        1. src1 = Matriz de referência à primeira imagem
        2. alpha = Intensidade da primeira imagem
        3. src2 = Matriz de referência à segunda imagem
        4. beta = Intensidade da segunda imagem 
        5. gamma = Valor escalar adiocinado a cada soma
        
Saída:

        1. Matriz de uma imagem mesclada

```python
imgTotal = cv.addWeighted(img1, 0.8, img2, 0.5, 0) # Misturando imagens
```

```python
fig = plt.figure(figsize=(20,5))
plt.imshow(imgTotal, cmap=plt.cm.gray)
```
