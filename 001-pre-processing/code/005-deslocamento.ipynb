## Translação da imagem
A operação de translação consiste em deslocar a imagem de posição

Imports

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

Lendo a imagem

```python
path = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\001-pre-processing\images\einstein.jpg'
imgOriginal = cv.imread(path, 0)
```

```python
print(imgOriginal)
```

Função float32([h,v,px], [h,v,px])

Entrada:
           
           1. [h,v,px] = horizontal, vertical, quantidade de pixels referente ao deslocamento
Saída:
           
           1. Nova imagem deslocada

```python
totalLinhas, totalColunas = imgOriginal.shape
print(totalLinhas, totalColunas)
```

```python
matriz = np.float32([[1,0,100], [0,1,100]])
imgDeslocada = cv.warpAffine(imgOriginal, matriz, (totalColunas, totalLinhas))
```

```python
fig = plt.figure(figsize=(10,8))
plt.imshow(imgDeslocada, cmap = plt.cm.gray)
```
