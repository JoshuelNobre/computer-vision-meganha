## Filtro Média
O filtro média é um filtro linear, classificado como passa-baixas. Esse tipo de filtro substitui cada pixel da imagem pelo valor médio da sua vizinhança

Imports

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

Lendo a imagem

```python
path = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\002-smoothing-filter\images'
```

```python
img = cv.imread(path + '\einstein.png')
```

Função blur(img, máscara)

Entrada:
        
        1. img = Imagem com a qual quero trabalhar 
        2. máscara = dimensão de máscara que será aplicada
       
Saída:
        
        1. Matriz referente a imagem tratada

```python
imgM = cv.blur(img, (7,7))
```

```python
fig = plt.figure(figsize=(20,50))
ax1 = fig.add_subplot(121)
plt.imshow(img)
plt.title("Imagem com ruído")

ax2 = fig.add_subplot(122)
plt.imshow(imgM)
plt.title("Imagem filtrada")

plt.show()
```
