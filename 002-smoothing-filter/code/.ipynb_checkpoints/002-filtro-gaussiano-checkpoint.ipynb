## Filtro Gaussiano
O filtro gaussianoé um filtro linear, passa-baixas. Apresenta bons resultados no tratamento de imagens com ruído gaussiano.

Imports

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

Lendo a imagem 

```python
path = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\002-smoothing-filter\images\formas.png'
```

```python
img = cv.imread(path)
```

Função GaussianBlur(img, máscara, suavização)

Entrada:

        1. img = imagem com qual quero trabalhar
        2. máscara = dimensão da máscara que será aplicada
        3. suavização = grau de suavização
        
Saída:

        1. matriz referente a imagem tratada

```python
imgGauss = cv.GaussianBlur (img, (5,5), 0)
```

```python
fig = plt.figure(figsize=(20,50))
ax1 = fig.add_subplot(121)
plt.imshow(img)
plt.title("Imagem original")

ax2 = fig.add_subplot(122)
plt.imshow(imgGauss)
plt.title("Imagem Gaussiana")

plt.show()
```
