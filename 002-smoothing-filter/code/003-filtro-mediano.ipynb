## Filtro mediano
O filtro mediano é do tipo não linear e apresenta bons resultados no tratamento do ruído do tipo "sal e pimenta".
É um tipo de técnica de tratamento de imagem capaz de suavizar preservando bordas e contornos da imagem.

Imports

```python
import cv2 as cv
from matplotlib import pyplot as plt 
%matplotlib inline
```

Lendo a imagem

```python
path = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\002-smoothing-filter\images\head.png'
```

```python
img = cv.imread(path)
```

Função medianBlur(img, intensidade)

Entrada:
        
        1. img = imagem que quero trabalhar
        2. intensidade = grau de suavização
        
Saída:

        1. matriz refente a imagem tratada

```python
imgMedian = cv.medianBlur(img, 5)
```

```python
fig = plt.figure(figsize=(20,50))

ax1 = fig.add_subplot(121)
plt.imshow(img)
plt.title("Imagem original")

ax2 = fig.add_subplot(122)
plt.imshow(imgMedian)
plt.title("Imagem com Mediana")

plt.show()
```
