## Detector de bordas - CANNY
O detector de bordas por canny é um dos mias eficientes algoritmos para detectar bordas em imagens. Esse tipo de detector de bordas deve respeitar três características principais

    1. O algoritmo deve ser capaz de identificar todas as bordas possíveis.
    2. Todas as bordas detctadas devem estar próximas das bordas originais.
    3. Bordas falsas não podem ser criadas, ou seja, cada borda deve ser definida uma única vez.

Imports

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

Lendo a imagem

```python
path = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\003-edge-filter\images'
img = cv2.imread(path + '\predio.jpg')
```

função Canny(img, int_detec1, int_detec2)

Entrada:
        
        1. img = Imagem com a qual queremos trabalhar em tons de cinza
        2. int_detec1 = Intensidade de detecção 1.
        3. int_detec2 = Intensidade de detecção 2.
        
Saída:
        
        1. Matriz referente a imagem com realce por Canny

```python
canny = cv2.Canny(img, 75, 100)

fig = plt.figure(figsize=(20,50))

ax1 = fig.add_subplot(121)
plt.imshow(img, cmap=plt.cm.gray)
plt.title('Imagem Original')

ax2 = fig.add_subplot(122)
plt.imshow(canny, cmap=plt.cm.gray)
plt.title('Operador Canny')

plt.show()
```
