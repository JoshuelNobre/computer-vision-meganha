### Filtro edgePreservingFilter

O filtro edgePreservingFilter, ou suavizao com preservao de bordas,  uma tcnica de processamento de imagem que suaviza rudos ou texturas enquanto mantm as bordas ntidas

Usado em muitas aplicacaes, por exemplo imagens mdicas ou de satlite, onde as bordas das imagens so caractersticas chave, portanto, devem ser preservadas ntidas e sem distores na suavizao de rudo. Filtro de preservao de borda so projetados para limitar automaticamente a suavizao em "bordas" em imagens

Entrada:

        1. dst = Imagem de sada
        2. flags = Filtros de preservao de borda. RECURS_FILTER(Filtragem Recursiva) = 1 e NORMCONV_FILTER(Convoluo Normalizada) = 2. Usar a opo RECURS_FILTER  cerca de 3,5x mais rpido que NORMCONV_FILTER. O NORMCONV_FILTER produz nitidez das bordas e  bom para aplicaes de estilizaes. Quando a nitidez no  desejvel e a velocidade  imporante, RECURS_FILTER deve ser usado.
        3. sigma_s = Valores entre 0 e 200. Controla o tamanho dos vizinhos para suavizao
        4. sigma_r = Valores entre 0 e 1. Controla como as cores diferentes na vizinhaa sero calculadas. Um sigma_r maior resulta em grandes regies de cor constante.

Desfoque da imagem usando a funo edgePreservingFilter

```python
edgePreservingImage = cv2.edgePreservingFilter(image, flags=2, sigma_s=50, sigma_r=0.4)

plt.figure(figsize=(10,8))
plt.imshow(edgePreservingImage)
plt.show()
```

### Filtro Stylization
O filtro de estilizao produz uma sada que parece que imagem foi pintada com aquarela.

Entrada:

        1.sigma_s = Valores entre 0 e 200. Controla o tamanho dos vizinhos para a suavizao
        2.sigma_r = Valores entre 0 e 1. Controla como as cores diferentes na vizinhana sero calculadas. Um sigma_r maior resulta em grandes regies de cor constante.

```python
cartoon_image = cv2.stylization(image, sigma_s=150, sigma_r=0.25)

plt.figure(figsize=(10,8))
plt.imshow(cartoon_image)
plt.show()
```
