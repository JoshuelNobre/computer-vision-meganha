### Filtro edgePreservingFilter

O filtro edgePreservingFilter, ou suavização com preservação de bordas, é uma técnica de processamento de imagem que suaviza ruídos ou texturas enquanto mantém as bordas nítidas

Usado em muitas aplicacações, por exemplo imagens médicas ou de satélite, onde as bordas das imagens são características chave, portanto, devem ser preservadas nítidas e sem distorções na suavização de ruído. Filtro de preservação de borda são projetados para limitar automaticamente a suavização em "bordas" em imagens

Entrada:

        1. dst = Imagem de saída
        2. flags = Filtros de preservação de borda. RECURS_FILTER(Filtragem Recursiva) = 1 e NORMCONV_FILTER(Convolução Normalizada) = 2. Usar a opção RECURS_FILTER é cerca de 3,5x mais rápido que NORMCONV_FILTER. O NORMCONV_FILTER produz nitidez das bordas e é bom para aplicações de estilizações. Quando a nitidez não é desejável e a velocidade é imporante, RECURS_FILTER deve ser usado.
        3. sigma_s = Valores entre 0 e 200. Controla o tamanho dos vizinhos para suavização
        4. sigma_r = Valores entre 0 e 1. Controla como as cores diferentes na vizinhaça serão calculadas. Um sigma_r maior resulta em grandes regiões de cor constante.

Desfoque da imagem usando a função edgePreservingFilter

```python
edgePreservingImage = cv2.edgePreservingFilter(image, flags=2, sigma_s=50, sigma_r=0.4)

plt.figure(figsize=(10,8))
plt.imshow(edgePreservingImage)
plt.show()
```

### Filtro Stylization
O filtro de estilização produz uma saída que parece que imagem foi pintada com aquarela.

Entrada:

        1.sigma_s = Valores entre 0 e 200. Controla o tamanho dos vizinhos para a suavização
        2.sigma_r = Valores entre 0 e 1. Controla como as cores diferentes na vizinhança serão calculadas. Um sigma_r maior resulta em grandes regiões de cor constante.

```python
cartoon_image = cv2.stylization(image, sigma_s=150, sigma_r=0.25)

plt.figure(figsize=(10,8))
plt.imshow(cartoon_image)
plt.show()
```
