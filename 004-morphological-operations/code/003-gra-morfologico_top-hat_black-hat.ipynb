## Gradiente morfológico, TOP HAT e BLACK HAT
    1. Gradiente morfológico

A operação chamada de gradiente morfológico é a diferença entre a operação de dilatação e rosão de uma iamgem. O resultado é uma imagem que representa a borda do objeto de interesse

    2. TOPHAT
A operação de TOPHAT consiste na subtração da versão morfologicamente aberta de uma imagem com a versão original. Esse tipo de operação é aplicada a fim de realçar objetos brilhantes em fundos escuros. Além dessa utilidade, ela também é aplicada com intuito de corrigir a variação de luminosidade em imagens, possibilitando que objetos representados em regiões mais escuras sejam realçados

    3. BLACK HAT
Essa operação consiste na subtração da imagem original por sua versão morfologicamente fechada.

```python
path = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\004-morphological-operations\images'
img1 = cv2.imread(path + "\moeda.png")
img2 = cv2.imread(path + "\esqueleto.tif")
```

### Elemento estruturante

```python
elem_str1 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
elem_str2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (21,21))

print(elem_str1)
print(elem_str2)
```

funcao morphologyEx(img, oper, elem_str)

Entrada:

        1. img = Imagem com a qual queremos trabalhar
        2. open = Operação que será realizada
        3. elem_estr = Elemento estruturante
        
Saída:
        
        1. Imagem nova

```python
imgProc1 = cv2.morphologyEx(img1, cv2.MORPH_GRADIENT, elem_str1)
imgProc2 = cv2.morphologyEx(img1, cv2.MORPH_TOPHAT, elem_str1)
imgProc3 = cv2.morphologyEx(img1, cv2.MORPH_BLACKHAT, elem_str1)

fig = plt.figure(figsize=(15,15))

ax1 = fig.add_subplot(141)
plt.imshow(img1)
plt.title("Imagem Original")

ax1 = fig.add_subplot(142)
plt.imshow(imgProc1)
plt.title("Gradiente Morfológico")

ax2 = fig.add_subplot(143)
plt.imshow(imgProc2)
plt.title("TOP HAT")

ax3 = fig.add_subplot(144)
plt.imshow(imgProc3)
plt.title("BLACK HAT")

plt.show()
```

```python
imgProc1 = cv2.morphologyEx(img2, cv2.MORPH_GRADIENT, elem_str2)
imgProc2 = cv2.morphologyEx(img2, cv2.MORPH_TOPHAT, elem_str2)
imgProc3 = cv2.morphologyEx(img2, cv2.MORPH_BLACKHAT, elem_str2)

fig = plt.figure(figsize=(15,15))

ax1 = fig.add_subplot(141)
plt.imshow(img2)
plt.title("Imagem Original")

ax1 = fig.add_subplot(142)
plt.imshow(imgProc1)
plt.title("Gradiente Morfológico")

ax2 = fig.add_subplot(143)
plt.imshow(imgProc2)
plt.title("TOP HAT")

ax3 = fig.add_subplot(144)
plt.imshow(imgProc3)
plt.title("BLACK HAT")

plt.show()
```
