```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\007-feature-extraction\images\hospital2.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

FUNÇÃO cv2.goodFeaturesToTrack(img, n, qual, k)

Entrada:

        1. img = imagem de entrada em tons de cinza
        2. n = números de cantos que deseja-se encontrar
        3. qual = nível de qualidade (valor entre 0 e 1)
        4. k = distância euclidiana mínima entre os cantos detectados

Saída:

        1. pontos de detecção de canto da imagem

```python
corners = cv2.goodFeaturesToTrack(gray, 10, 0.05, 0.25)

for item in corners:
    x,y = item[0]
    cv2.circle(img, (int(x),int(y)), 4, (0,0,255), -1)

fig = plt.figure(figsize=(10,8))
plt.imshow(img)
plt.show()
```
