### Tesseract
**image_to_boxes:** Retorna o resultado do reconhecimento de caracteres através de caixas

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

```python
path = r'C:\Users\shuel\Documents\Workspace\computer-vision\computer-vision-meganha\010-pytesseract\images\teste_tesseract.png'
```

```python
img = cv2.imread(path)
# Como o openCV trabalha em BGR e o pytesseract RGB, iremos fazer essa conversão
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Mostar a imagem
plt.imshow(img)
plt.show
```

```python
box = pytesseract.image_to_boxes(img)
print(box)
```

```python
h_img, w_img, _ = img.shape
print(h_img)
print(w_img)
```

```python
for b in box.splitlines():
    b = b.split(' ')
    # print(b)
    x,y,w,h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
    cv2.rectangle(img, (x, h_img-y), (w, h_img-h), (50, 50, 255), 2)
    cv2.putText(img, b[0], (x, h_img-y + 25), cv2.FONT_HERSHEY_SIMPLEX, 1, (50,50,255), 2)
    
plt.imshow(img)
plt.show()
```
