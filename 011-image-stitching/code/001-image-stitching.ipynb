# ORB

*   Boa alternativa para o SIFT e o SURF em relao ao custo computacional e performaces de macths. 
*   O ORB  basicamente a fuso entre a deteco de keypoints do FAST e as descries feitas pelo BRIEF. Primeiramente o FAST encontra os keypoints e ento encontra os melhores N pontos. Despois BRIEF avalia as respectivas correspondncias encontradas pelos pontos.
*    O uso mais comum do ORB  passar o nFeatures, responsvel pelo mx. nmeros das features (default = 500) e o WTA_K que encontra a orientao das features encontras (default = NORM_HAMMING)


*  Parmetros:

1.  **nfeatures** O nmero mximo de recursos a serem retidos.

2.  **scaleFactor** Taxa de decimao da pirmide, maior que 1. scaleFactor == 2 significa a pirmide clssica, onde cada prximo nvel tem 4x menos pixels do que o anterior, mas esse fator de grande escala degradar drasticamente as pontuaes correspondentes. Por outro lado, muito prximo a 1 fator de escala significar que, para cobrir determinada faixa de escala, voc precisar de mais nveis de pirmide e, portanto, a velocidade de deteco sofrer.

3.  **nlevels** O nmero de nveis da pirmide.

4.  **edgeThreshold** Esse  o tamanho da borda em que os recursos no so detectados. Deve corresponder aproximadamente ao parmetro patchSize.

5.  **firstLevel** O nvel da pirmide na qual colocar a imagem de origem. As camadas anteriores so preenchidas com a imagem de origem aumentada.

6.  **WTA_K** O nmero de pontos que produzem cada elemento do descritor BRIEF orientado. O valor padro 2 significa o BRIEF em que pegar um par de pontos aleatrios e comparar seus brilhos, para obter uma resposta 0/1. Outros valores possveis so 3 e 4. Por exemplo, 3 significa que pegamos 3 pontos aleatrios ( claro, essas coordenadas de pontos so aleatrias, mas so geradas a partir da semente predefinida, de modo que cada elemento do descritor BRIEF  calculado deterministicamente a partir de o retngulo de pixels), encontre o ponto de brilho mximo e o ndice de sada do vencedor (0, 1 ou 2). Essa sada ocupar 2 bits e, portanto, precisar de uma variante especial da distncia de Hamming, indicada como NORM_HAMMING2 (2 bits por compartimento). Quando WTA_K = 4, tomamos 4 pontos aleatrios para calcular cada compartimento (que tambm ocupar 2 bits com valores possveis 0, 1, 2 ou 3).

7.  **scoreType** O HARRIS_SCORE padro significa que o algoritmo Harris  usado para classificar os recursos (a pontuao  gravada em KeyPoint::score e  usada para manter os melhores recursos de nfeatures); FAST_SCORE  um valor alternativo do parmetro que produz pontos-chave um pouco menos estveis, mas  um pouco mais rpido de calcular.

8.  **patchSize** Tamanho usado pelo descritor BRIEF orientado.

9.  **fastThreshold** encontrar o Threshold de maneira rpida.



```python
detect = cv2.ORB_create()

kp1, feat1 = detect.detectAndCompute(img1_gray, None)
kp2, feat2 = detect.detectAndCompute(img2_gray, None)

fig = plt.figure(figsize=(20,50))

ax1 = fig.add_subplot(121)
ax1.imshow(cv2.drawKeypoints(img1, kp1, None, color=(0,255,0)))
ax1.set_xlabel('Imagem 1', fontsize=14)

ax2 = fig.add_subplot(122)
ax2.imshow(cv2.drawKeypoints(img2, kp2, None, color=(0,255,0)))
ax2.set_xlabel('Imagem 2', fontsize=14)

plt.show()
```

# BFMatcher

O brute-Force matcher  simples. Ele pega o descritor de um recurso/feature da primeira imagem e combina com todos os outros recursos/features da segunda imagem usando algum clculo de distncia. E o mais prximo  retornado.

Para o BF matcher, so necessrios dois parmetros opcionais. O primeiro  o **normType**. Especifica a medio da distncia a ser usada. Por padro,  **cv2.NORM_L2**.  bom para SIFT, SURF etc (cv2.NORM_L1 tambm est l). Para descritores baseados em cadeias binrias como ORB, BRIEF, BRISK etc, **cv2.NORM_HAMMING** deve ser usado, que usou a distncia de Hamming como medida. Se o ORB estiver usando WTA_K == 3 ou 4, cv2.NORM_HAMMING2 deve ser usado.

O segundo parmetro  a varivel booleana, **crossCheck**, que  falsa por padro. Se for verdade, o Matcher retornar apenas as correspondncias com o valor (i, j), de modo que o i-simo descritor no conjunto A tenha o j-simo descritor no conjunto B como a melhor correspondncia e vice-versa. Ou seja, os dois recursos nos dois conjuntos devem coincidir. Ele fornece resultados consistentes e  uma boa alternativa ao teste de proporo proposto por D.Lowe no papel SIFT.

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
best_matches = bf.match(feat1, feat2)

raw_matches = sorted(best_matches, key=lambda x:x.distance)
print('Matches:', len(raw_matches))

fig = plt.figure(figsize=(20, 8))

img3 = cv2.drawMatches(img1, kp1, img2, kp2, best_matches[:100], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

plt.imshow(img3)
plt.show()
```

# findHomography

Encontrar a pespectiva entre dois planos.

Parmetros:

**srcpoints** - Coordenadas dos pontos da imagem original.
**dstPoints** - Coordenadas dos pontos da imagem alvo.
**method** - Mtodo usada para computar a homografia da imagem.  possvel passar trs mtodos: 
1.    0 - Um mtodos regular usado em todos os pontos.
2.    RANSAC - Mtodos robusto baseado em RANSAC
3.    LMEDS -  Menor mediana

**ransacReprojThreshold** - Erro mximo permitido de reprojeo para tratar um par de pontos em inliers (mtodo usado apenas quando  setado o RANSAC). Ou seja, se:

||dstPoints - convertPointsHomogeneous(H*srcPoints)|| > ransacReprojThreshold

ento o ponto em questo  considerado um desvio. Se srcPoints e dstPoints forem medidos em pixels, geralmente faz sentido definir esse parmetro em algum lugar no intervalo de 1 a 10.


Ambos os mtodos, RANSAC e LMeDS, etstam muitos subconjuntos aleatrios diferentes dos pares de pontos correspondentes (de quatro pares cada), estimam a matriz de homografia usando esse subconjunto e um algoritmo simples de mnimos quadrados e depois calculam a qualidade/qualidade da homografia computada (que  o nmero de inliers para o RANSAC ou o erro mdio de re-projeo para os LMeDs). O melhor subconjunto  ento usado para produzir a estimativa inicial da matriz de homografia e a mscara de inliers / outliers.

Independentemente do mtodo, robusto ou no, a matriz de homografia computada  refinada ainda mais (usando inliers apenas no caso de um mtodo robusto) com o mtodo Levenberg-Marquardt para reduzir ainda mais o erro de re-projeo.

O mtodo RANSAC pode lidar com praticamente qualquer proporo de outliers, mas precisa de um limite para distinguir inliers de outliers. O mtodo LMeDS no precisa de nenhum limite, mas funciona corretamente apenas quando h mais de 50% de inliers. Por fim, se no houver discrepncias e o rudo for bastante pequeno, use o mtodo padro (mtodo = 0)




```python
src_pts = np.float32([kp1[m.queryIdx].pt for m in raw_matches]).reshape(-1,1,2) 
dst_pts = np.float32([kp2[m.trainIdx].pt for m in raw_matches]).reshape(-1,1,2) 

H, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5, 0)

h, w = img1_gray.shape
pts = np.float32([[0,0],[0,h-1], [w-1, h-1], [w-1,0]]).reshape(-1,1,2)
dst = cv2.perspectiveTransform(pts, H)

img_final = cv2.polylines(img2, [np.int32(dst)], True, 255,1, cv2.LINE_AA)

fig = plt.figure(figsize=(20,8))
plt.imshow(img_final)
plt.show()
```
